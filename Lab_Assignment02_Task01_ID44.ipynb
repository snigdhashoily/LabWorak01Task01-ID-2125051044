```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
# Step 1: Provide the URL of the image
image_url = 'https://cdn.shopify.com/s/files/1/0274/9503/9079/files/20220211142754-margherita-9920_5a73220e-4a1a-4d33-b38f-26e98e3cd986.jpg?v=1723650067'
```

```python
# Step 2: Load the image from the URL
def load_image_from_url(url):
    # Open the URL and read the image into memory
    with urllib.request.urlopen(url) as response:
        img_array = np.array(bytearray(response.read()), dtype=np.uint8)
        img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
    return img
```

```python
# Step 3: Rotate the image by 180 degrees
def rotate_image(img, angle=180):
    # Rotate the image by the specified angle (in degrees)
    rotated_img = cv2.rotate(img, cv2.ROTATE_180)  # 180 degrees rotation
    return rotated_img
```

```python
# Step 4: Load the image from the URL
img = load_image_from_url(image_url)
```

```python
# Step 5: Rotate the image by 180 degrees
rotated_img = rotate_image(img)
```

```python
# Step 6: Display the original and rotated images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Rotated Image (180 degrees clockwise)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image (180 degrees)')
plt.axis('off')

plt.show()
```
